Ustalenie zmiennych

```python
import pandas as pd
import nltk

nltk.download('punkt')
dane = pd.read_csv('spam_ham_dataset.csv')

slowa = {}
tokenizer = nltk.RegexpTokenizer(pattern='[a-zA-Z]+')
for i in range(len(dane)):
    mail = dane['text'][i]
    mail_slowa = tokenizer.tokenize(mail)
    if dane['label_num'][i]==0:
        for j in range(len(mail_slowa)):
            if mail_slowa[j].lower() in slowa:
                slowa[mail_slowa[j].lower()] += 1
            else:
                slowa[mail_slowa[j].lower()] = 1
    if dane['label_num'][i]==1:
        for j in range(len(mail_slowa)):
            if mail_slowa[j].lower() in slowa:
                slowa[mail_slowa[j].lower()] -= 1
            else:
                slowa[mail_slowa[j].lower()] = -1

posortowane = sorted(slowa.items(), key=lambda x: x[1])
print('spam')
for i in range(10):
    print(posortowane[i])
print('nie spam')
for i in range(10):
    print(posortowane[len(posortowane)-1-i])
```

Obliczenie zmiennych

```python
zmienne = {'Długość':[],'CapsLock_śr':[],'CapsLock_Max':[],'CapsLock_Sum':[]}
CapsTokenizer = nltk.RegexpTokenizer(pattern='[A-Z]+')
for i in range(10):
    zmienne[posortowane[i][0]]=[]
for i in range(10):
    zmienne[posortowane[len(posortowane)-i-1][0]]=[]
for i in range(len(dane)):
    mail = dane['text'][i]
    mail_slowa = tokenizer.tokenize(mail)
    for j in range(10):
        zmienne[posortowane[j][0]].append(0)
    for j in range(10):
        zmienne[posortowane[len(posortowane)-j-1][0]].append(0)
    for j in range(len(mail_slowa)):
        if mail_slowa[j].lower() in zmienne:
            zmienne[mail_slowa[j].lower()][i]+=1
    zmienne['Długość'].append(len(mail))
    mail_Caps = CapsTokenizer.tokenize(mail)
    Sum = 0
    Max = 0
    for j in range(len(mail_Caps)):
        Sum += len(mail_Caps[j])
        if len(mail_Caps[j])>Max:
            Max = len(mail_Caps[j])
    zmienne['CapsLock_Sum'].append(Sum)
    zmienne['CapsLock_śr'].append(Sum / len(mail_Caps))
    zmienne['CapsLock_Max'].append(Max)

csv = pd.DataFrame(zmienne)
csv.to_csv('dane.csv')
```
